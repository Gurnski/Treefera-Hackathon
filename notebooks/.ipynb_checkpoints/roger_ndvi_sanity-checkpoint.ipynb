```python
from pathlib import Path
import xarray as xr

root = Path(r"C:\Users\danie\Desktop\Treefera Hackathon\data\sentinel2\B_grinnell_iowa\2017-01-01_2025-12-31")
zarr_path = root / "cube.zarr"

print("Root exists:", root.exists())
print("Zarr exists:", zarr_path.exists())
print("Preview exists:", (root / "preview_median.tif").exists())

ds = xr.open_zarr(zarr_path, consolidated=False)

print(ds)
print("\nData variables:", list(ds.data_vars))
print("\nCoordinates:", list(ds.coords))
print("\nDimensions:", ds.dims)

if "time" in ds.coords:
    print("\nTime range:", ds.time.min().values, "to", ds.time.max().values)
    print("Time count:", ds.sizes["time"])
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

red = ds["B04"].astype("float32")
nir = ds["B08"].astype("float32")

ndvi = (nir - red) / (nir + red)
ndvi = ndvi.where(np.isfinite(ndvi))

spatial_dims = [d for d in ["x", "y"] if d in ndvi.dims]
print("Spatial dims:", spatial_dims)

mean_ndvi = ndvi.mean(dim=spatial_dims, skipna=True).compute()

df = mean_ndvi.to_dataframe(name="ndvi").reset_index()
display(df.head())
display(df.tail())

plt.figure(figsize=(12, 5))
plt.plot(df["time"], df["ndvi"], marker="o", linewidth=1.5)
plt.axvline(pd.Timestamp("2021-01-01"), linestyle="--", label="Cover crops from 2021")
plt.axvline(pd.Timestamp("2023-01-01"), linestyle="--", label="No-till from 2023")
plt.title("Grinnell / Roger's Farm area — whole-scene mean NDVI")
plt.ylabel("NDVI")
plt.xlabel("Date")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```

```python
from pathlib import Path

project_root = Path(r"C:\Users\danie\Desktop\Treefera Hackathon")
data_root = project_root / "data"

wanted_exts = {".geojson", ".json", ".shp", ".gpkg", ".csv"}
keywords = ["field", "fields", "polygon", "boundary", "boundaries", "parcel", "roger", "control", "neighbour", "neighbor", "farm"]

matches = []
for p in data_root.rglob("*"):
    if p.is_file() and p.suffix.lower() in wanted_exts:
        name = str(p).lower()
        if any(k in name for k in keywords):
            matches.append(p)

print(f"Found {len(matches)} possible boundary/control files:")
for p in matches[:100]:
    print(p)
```
